# 2. Data Preprocessing

Na etapa de pré-processamento vamos preparar nossos dados para os modelos de clusterização. Aqui faremos:
1. Normalização dos dados
2. Data Wrangling para criar as métricas de RFM

```python
# Antes de Normalizarmos nossos dados vamos criar as métricas RFM (Recency, Frequency, Monetary)

# Supondo que usaremos InvoiceDate como data das compras
# Calculando a recência (em dias) a partir da data mais recente
recent_date = df_filtered['InvoiceDate'].max()
df_filtered['Recency'] = (recent_date - df_filtered['InvoiceDate']).dt.days
```

```python
#Dessa forma, o DF fica da seguinte forma
df_filtered.head()
```

```python
# Agrupando por 'CustomerID' e calculando a frequência e o valor monetário
rfm_table = df_filtered.groupby('CustomerID').agg({
    'Recency': 'min',  # Usamos min para obter a recência mais recente
    'InvoiceNo': 'count',  # Contamos o número de compras para obter a frequência
    'UnitPrice': 'sum'  # Somamos os preços unitários para obter o valor monetário total
})

# Renomeando as colunas para Recency, Frequency e Monetary
rfm_table.rename(columns={
    'InvoiceNo': 'Frequency',
    'UnitPrice': 'Monetary'
}, inplace=True)

rfm_table.head()
```

```python
# Adicionando os valores RFM ao dataframe original
df_filtered_rfm = pd.merge(df_filtered, rfm_table, on='CustomerID')

# Removendo as colunas duplicadas de Recency
df_filtered_rfm.drop(columns ='Recency_x', inplace=True)
df_filtered_rfm.rename(columns={
    'Recency_y': 'Recency',
}, inplace=True)
# Exibindo o dataframe com as colunas RFM adicionadas
df_filtered_rfm.head()
```

Agora que temos os dados bem estruturados e estudados, podemos normalizá-los.

O método que será usado para normalizar será o de Standard Scaler.

No geral, a padronização z-score é uma escolha sólida para normalizar os dados RFM, especialmente quando se considera a robustez em relação aos outliers e a interpretação direta dos resultados. No entanto, é importante estar ciente das limitações e considerar a adequação da padronização z-score para os seus dados específicos. São elas:

**Prós:**

1. **Interpretabilidade:** Os coeficientes resultantes da padronização z-score são interpretáveis diretamente em termos de desvios padrão da variável original. Isso facilita a interpretação dos resultados da clusterização.

2. **Robustez:** A padronização z-score é robusta em relação aos outliers. Ela usa a média e o desvio padrão dos dados, que não são afetados por valores extremos.

3. **Preserva a forma da distribuição:** A padronização z-score não altera a forma da distribuição dos dados, apenas os escala e os centraliza em torno de zero.


**Contras:**

1. **Sensibilidade ao tamanho da amostra:** Se a amostra for pequena, a média e o desvio padrão podem ser sensíveis a outliers ou variações extremas nos dados.

2. **Assume normalidade:** A padronização z-score assume que os dados seguem uma distribuição normal. Se os dados forem muito assimétricos ou não seguirem uma distribuição normal, a padronização z-score pode não ser a melhor escolha.

3. **Interpretação mais complexa com muitas variáveis:** Se houver muitas variáveis no modelo, a interpretação dos coeficientes padronizados pode se tornar mais complexa.

```python
# Selecionar as colunas Recency, Frequency e Monetary
rfm_subset = rfm_table[['Recency', 'Frequency', 'Monetary']]

# Inicializar o scaler
scaler = StandardScaler()

# Aplicar a padronização z-score
rfm_standardized = scaler.fit_transform(rfm_subset)

# Criar um novo DataFrame com os dados padronizados
rfm_standardized_df = pd.DataFrame(rfm_standardized, columns=['Recency', 'Frequency', 'Monetary'], index=rfm_table.index)
```

# Data Modelling

O primeiro passo para a modelagem é identificar qual modelo faz mais sentido para nosso dataset. Para começar a modelagem usarei o KMeans devido à sua simplicidade, eficiência e escalabilidade

No entanto, o KMeans é sensível a outliers, portanto se o KMeans não atender às expectativas considerarei o uso do DBSCAN, que é mais robusto contra outliers e não requer a definição prévia do número de clusters.

Essa abordagem garantirá que eu escolha o melhor algoritmo para classificar os clientes.

```python
# O primeiro passo será usar o método elbow para determinar o melhor número de clusters quando usarmos o Kmeans
# faremos isso com o kelbow_visualizer
# Definindo o tamanho do plot
plt.figure(figsize=(10, 6))

# Criando e visualizando o gráfico do cotovelo
kelbow_visualizer = KElbowVisualizer(KMeans(n_init=10), k=(2,10))
kelbow_visualizer.fit(rfm_standardized_df)
kelbow_visualizer.show()
```

Através do método do cotovelo é notável que podemos escolher a partir de 4 clusters para nosso método, porém 5 clusters não apenas diminuem o distortion_score como reduz o fit time pela metade.

```python
# Outra maneira de identificar o n° ótimo de clusters é pelo silhouette_visualizer, portanto faremos ele para os clusters de 3 até 6

@interact(k=(2,6))

def plot_silhouette_visualizer_(k):
  plt.figure(figsize=(8, 5))
  silhouette_visualizer(KMeans(n_clusters=k, n_init='auto'),rfm_standardized_df);
  plt.show()
```

Uma outra forma de visualizarmos diferentes métricas para determinar o melhor n° de Clusters é criando uma tabela com as métricas mais relevantes para nossa análise. Da seguinte forma:

```python
cluster_metrics = silhouette_score, davies_bouldin_score, calinski_harabasz_score
cluster_metrics_results= []

for k in range (2,9):
  model= KMeans(n_clusters= k, n_init='auto')
  labels=model.fit_predict(rfm_standardized_df)
  cluster_results_dict = {'n_clusters': k}
  cluster_results_dict['inertia']=model.inertia_
  for metric in cluster_metrics:
    cluster_results_dict[metric.__name__]= metric(rfm_standardized_df, labels)
  cluster_metrics_results.append(cluster_results_dict)

(
  pd.DataFrame(cluster_metrics_results)
  .set_index('n_clusters')
  .style.background_gradient(cmap='Greens', subset='inertia')
  .highlight_max(subset=['silhouette_score','calinski_harabasz_score'], color='green')
  .highlight_min(subset='davies_bouldin_score', color='green')
)
```

Como obtivemos diferentes resultados conforme a métrica utilizada, vamos avaliar o que cada uma quer dizer:

1. **Inertia**: Mede a soma das distâncias quadradas entre cada ponto e o centróide do cluster, sendo útil para avaliar a compactação dos clusters. Quanto menor o valor, melhor, pois indica que os pontos estão mais próximos dos centróides dos clusters.

2. **Silhouette_score**: Calcula o quão bem cada ponto se ajusta ao seu próprio cluster em comparação com outros clusters, fornecendo uma medida da separação dos clusters. Valores próximos de 1 indicam que os pontos estão bem ajustados aos seus próprios clusters e mal ajustados aos clusters vizinhos.

3. **Davies_bouldin_score**: Avalia a dispersão entre os clusters, onde um valor menor indica uma melhor separação entre os clusters.Quanto menor o valor, melhor, indicando uma melhor separação entre os clusters.

4. **Calinski_harabasz_score**: Avalia a relação entre a dispersão dentro dos clusters e a dispersão entre os clusters, sendo maior quando os clusters estão densos e bem separados. Valores mais altos indicam uma melhor separação entre os clusters.

Agora que temos as métricas, precisamos decidir quantos clusters usaremos.

Usando o método Elbow, a partir de 4 clusters a distorção já cai consideravelmente e com 5 clusters temos um custo computacional bem menor.

Pelas métricas de silhouette_score e davies_bouldin_score, o número 4 continua aparecendo com um bom número.

No coeficiente de calinski_harabasz_score só temos os melhores valores a partir de 7 clusters, ou seja, só teremos clusters densos e muito bem separados acima de 5 clusters, como podemos ver no cotovelo abaixo em que plotei a curva do coeficiente de Calinski-Harabasz.

```python
model = KMeans(n_init='auto')
visualizer = KElbowVisualizer(
    model, k=(4,12), metric='calinski_harabasz', timings=False, locate_elbow=False
)

visualizer.fit(rfm_standardized_df)
visualizer.show()
```

Vamos então plotar como ficam os clusters se dividirmos em 4 clusters o nosso dataset

```python
model = KMeans(n_clusters=4, n_init='auto')
model.fit(rfm_standardized_df)

# Atribua as etiquetas dos clusters aos dados
centroids = model.cluster_centers_
labels= model.labels_
inertia= model.inertia_

# Visualize output
fig = px.scatter_3d(rfm_standardized_df, x='Recency', y='Frequency', z='Monetary', color=labels, template='plotly_dark');
fig.show()
```

Ao que tudo indica, 4 clusters parecem ser a solução, apesar de apresentar um baixo coeficiente de Calinski-Haranasz.

Portanto, como forma de estudar mais a fundo, vamos aplicar o DBScan, pois também é um modelo compatível com nosso caso.

Antes de realizar o fit do modelo DBSCAN, podemos encontrar os melhores hiperparâmetros para o modelo.

Uma das maneiras de se visualizar isso é através do plot de um gráfico, que farei usando a biblioteca Plotly para permitir melhor exploração dos dados

```python
# Primeiramente usaremos o método de NearestNeighbors para descobrir qual o epsilon ideal para nosso modelo DBscan
# Calculando os vizinhos mais próximos
nn_model = NearestNeighbors(n_neighbors=2).fit(rfm_standardized_df)
distances, indices = nn_model.kneighbors()
distances = np.sort(distances, axis=0)

# Criando o gráfico
fig = go.Figure()

# Adicionando a linha ao gráfico
fig.add_trace(go.Scatter(x=np.arange(len(distances[:, 1])), y=distances[:, 1], mode='lines'))

# Personalizando o layout do gráfico
fig.update_layout(
    title="Distâncias para os vizinhos mais próximos",
    xaxis_title="Índices",
    yaxis_title="Distâncias",
    xaxis=dict(
        tickmode='linear',
        tick0=0,
        dtick=500
    ),
    yaxis=dict(
        range=[0, 2],
        tickmode='linear',
        tick0=0,
        dtick=0.2
    )
)

# Exibindo o gráfico
fig.show()
```

Agora que plotamos o gráfico identificamos que o epsilon ideal é em torno de
 0.6

 Para seguirmos, podemos usar o hiperparâmetro de min_values como x*(n° de colunas), ou seja Minimum samples = 6

```python
# Dessa forma, podemos começar nossa modelagem

# Defina os parâmetros do DBSCAN
epsilon = 0.6
min_samples = 6

# Crie uma instância do DBSCAN
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)

# Ajuste o modelo aos dados
dbscan.fit(rfm_standardized_df)

# Atribua as etiquetas dos clusters aos dados
cluster_labels = dbscan.labels_

# Visualize output
fig = px.scatter_3d(rfm_standardized_df, x='Recency', y='Frequency', z='Monetary', color=cluster_labels, template='plotly_dark')
fig.show()
```

Através da visualização acima, podemos identificar que cada modelo trouxe uma sugestão.

O Modelo KMeans retorna que o ideal é ter 4 clusters ou mais

Enquanto que o DBscan retorna entre 2 e 3 clusters, dependendo do epsilon usado.

Adotarei o modelo Kmeans com numero de 4 clusters, e o motivo será explicado a seguir.
